Project 1 — Stock Market Analysis using NumPy
Problem Statement

A financial analyst wants to analyze stock prices for a company over multiple days and calculate:

    Daily price changes
    
    Average price
    
    Maximum & minimum price
    
    Best day to buy and sell
    
    Moving average

Concepts Covered

    NumPy arrays
    
    Statistical operations
    
    Array slicing
    
    Aggregation functions

Dataset
    Daily stock prices (10 days)

```python
import numpy as np

#Stock Prices for 10 days
stockPrices=[120, 125, 123, 130, 128, 135, 140, 138, 142, 145]
prices=np.array(stockPrices)
print("Stock Prices:", prices)
```

```python
# Average price
avg_price = np.mean(prices)
print("Average Price: ",avg_price)
```

```python
# Maximum and Minimum price
print("Max Price:", np.max(prices))
print("Min Price:", np.min(prices))
```

```python
# Daily price change
daily_change = np.diff(prices)
print("Daily Change:", daily_change)
```

```python
# Best day to buy
buy_day = np.argmin(prices) #returns the index (position) of the minimum value in a NumPy array along a specified axis. 
print("Best Day to Buy:", buy_day)
```

```python
# Best day to sell
sell_day = np.argmax(prices)
print("Best Day to Sell:", sell_day)
```

```python
'''
The numpy.convolve function in Python performs a discrete, linear convolution of two one-dimensional arrays (sequences),
which is a common operation in signal processing and data analysis. 
It combines two sequences to create a third sequence representing their interaction.
'''
# Moving Average (3 days)
onesArr=np.ones(3)/3
print(onesArr)
moving_avg = np.convolve(prices, np.ones(3)/3, mode='valid') 
print("3 Day Moving Average:", moving_avg)
```
