```python
#union of two sets
x={1,2,4,5,6}
y={5,6,7,8,9,10}
print(x|y)
print(x.union(y))
print(x)
print(y)
```

```python
#intersection of two sets
x={1,2,4,5,6}
y={5,6,7,8,9,10}
print(x&y)
print(x.intersection(y))
```

```python
#difference of two sets
x={1,2,4,5,6}
y={5,6,7,8,9,10}
print(x-y)
print(x.difference(y))
```

```python
#symmetric_difference of two sets
x={1,2,4,5,6}
y={5,6,7,8,9,10}
print(x^y)
print(x.symmetric_difference(y))
```
