What is NumPy?
NumPy (Numerical Python) is the core Python library for scientific computing and numerical operations.
It provides:
Fast multidimensional arrays
Mathematical functions
Linear algebra operations
Random number generation
Matrix operations
NumPy is written in C, which makes it much faster than Python lists.

Why NumPy Instead of Python Lists?
| Feature                 | Python List | NumPy Array |
| ----------------------- | ----------- | ----------- |
| Speed                   | Slow        | Very Fast   |
| Memory                  | High        | Efficient   |
| Mathematical Operations | Difficult   | Very Easy   |
| Multi-dimensional       | Limited     | Built-in    |

```python
'''
    Mathematical Operations on Arrays

NumPy supports vectorized operations.
'''
#Addition
import numpy as np

a = np.array([10,20,30])
b = np.array([1,2,3])
c = a + b

print(c)
#Subtraction
print(a-b)
#Multiplication
print(a*b)
#Division
print(a/b)
```

```python
#Matrix Multiplication
import numpy as np

a = np.array([[1,2],
              [3,4]])
b = np.array([[5,6],
              [7,8]])
print(a)
print(b)
print('*******************')
print(a*b)
print('*******************')
result = np.dot(a,b)
print(result)
```

```python
import numpy as np

# Generate a 3x2 2D array with random floats between 0 and 1
# arr1 has 2 columns
arr1 = np.random.rand(3, 2)
print(f"Array 1 shape: {arr1.shape}")

# Generate a 2x4 2D array with random floats between 0 and 1
# arr2 has 2 rows
arr2 = np.random.rand(2, 4)
print(f"Array 2 shape: {arr2.shape}")

# Perform the dot product (matrix multiplication)
# The result will be a 3x4 array
dot_product_result = np.dot(arr1, arr2)
print(f"Dot product result shape: {dot_product_result.shape}")
print("Dot product result array:\n", dot_product_result)
```

```python
'''
Key Compatibility Rule
For matrix multiplication (which is what numpy.dot() performs with 2D arrays), the inner dimensions must match: 
If arr1 has shape (n, k) and arr2 has shape (k, m), the operation is valid.
The resulting array will have the shape (n, m).
If the inner dimensions (k in the example above) do not match, NumPy will raise a ValueError. 
'''
# Using the @ operator (preferred for matrix multiplication) -(Python 3.5+ preferred)
matrix_product_result = arr1 @ arr2
matrix_product_result
```

```python
'''
Functions that Apply to Each Element
NumPy applies functions element-wise.
'''
import numpy as np

arr = np.array([1,2,3,4])
print(np.square(arr)) #Square of each element
print(np.sqrt(arr)) #Square Root 
print(np.exp(arr)) #Exponential of each element of an array
print(np.log(arr)) # Natural logarithm (base e)
print(np.log10(arr)) ## Base 10 logarithm
print(np.log2(arr))# Base 2 logarithm
```

```python
'''
Retrieve Specific Parts of an Array
This is called Indexing and Slicing.
'''
import numpy as np
#index---------> 0 1  2   3
arr = np.array([10,20,30,40])
print(arr[0])
print(arr[2])
print("*************************")
#2D Array
arr = np.array([[10,20,30],
                [40,50,60]])
'''
         Col   Col   Col
          0    1      2
         (0,0) (0,1)  (0,2)
row 0    10    20    30

         (1,0) (1,1)  (1,2)  
row 1    40    50    60
'''
print(arr[1,1])
```

```python
#Row selection
print(arr[1])
#Column Selection
print(arr[:,1])
print('****************************')
#Slicing
print(arr[0:2,1:3])
```

```python
#Iterating Over Arrays
#Iterating 1D Array
import numpy as np

arr = np.array([10,20,30])

for x in arr:
    print(x)

for index,value in enumerate(arr):
    print(index," index contains element whose value is ",value)



```

```python
#2d array
arr_2d = np.array([[1,2,3],[3,4,5]])

for row in arr_2d:
    print(row)

for indices, value in np.ndenumerate(arr_2d):
    print(f"Indices {indices}: Value {value}")

'''
Key Differences
enumerate(): Provides a single, flat integer index, regardless of the array's dimensions.
numpy.ndenumerate(): Provides a tuple of indices corresponding to the element's exact position in a 
multidimensional array (e.g., (row, column) for a 2D array).
'''
```

```python
#Iterating Each Element
'''
np.nditer in Python is a multidimensional iterator designed to loop through elements of arrays with 
any number of dimensions, including 2D arrays.
The main purpose of np.nditer is to provide a consistent and efficient way to iterate over every scalar element of
an array using a single for loop, which removes the need for writing nested loops for multi-dimensional arrays.
'''
import numpy as np

arr = np.array([10,20,30])

for x in np.nditer(arr):
    print(x)


print("*********************************************")
import numpy as np
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
for x in np.nditer(arr_2d):
    print(x)
```
