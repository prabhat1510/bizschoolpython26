```python
#Demonstrating the use of multiple line comments
def multiply(x,y):
    '''
        This function takes two arguments and performs
        multiplication.
        It can take integer value as wells as decimal values.
        It returns result after multiplying these numbers
    '''
    result = x*y
    return result
#Call the function
print(multiply.__doc__)
res = multiply(15.10,10.00)
print(res)
```

```python
#Mixed Data type
mixedValues=[2,4,'Kiwi','Banana',15.60]
print(type(mixedValues))
print(mixedValues)
```

```python
'''
index->      0   1   2        3       4
             [2, 4, 'Kiwi', 'Banana', 15.6]
'''
print(mixedValues[3])
print(mixedValues[1])
print(mixedValues[4])
#print(mixedValues[5])
#print(mixedValues.len)
print(len(mixedValues))
print(mixedValues[1:3]) # 1 is start element index and 3 is last element index not included
print(mixedValues[1:4])
print(mixedValues[1:]) #starting from index pos 1 till last element in the list
```
