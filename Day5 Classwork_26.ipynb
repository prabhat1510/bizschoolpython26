```python
# Read the file using a while loop
with open('D:\\bizschoolpython\\2026\\story.txt', 'r') as myfile:
    next_line = myfile.readline()
    while next_line != "":
        # Process the line (e.g., print it after stripping the newline character)
        print(next_line.rstrip()) 
        # Read the next line
        next_line = myfile.readline()

# The file is automatically closed when the 'with' block is exited.
```

```python
import re

phones = [
    "(9988) 654-3210",
    "987-654-3210",
    "987 654 3210",
    "+91 98765-43210",
    "987.654.3210",
    " 9876543210 ",
    "(080) 2345 6789"
]

cleaned_numbers = []

for phone in phones:
    # Remove everything except digits
    digits_only = re.sub(r'\D', '', phone)
    cleaned_numbers.append(digits_only)

# Print result
for number in cleaned_numbers:
    print(number)
```
